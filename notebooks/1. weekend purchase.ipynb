go on with deriving the dif values

```python
difs=[]
for i in range(1,55):
    ID=i
    store_one=pd.DataFrame()
    store_one=df[df['StoreID']==ID]
    sat=store_one.groupby('day_of_week')['Sales'].sum()[5]
    sun=store_one.groupby('day_of_week')['Sales'].sum()[6]
    mon=store_one.groupby('day_of_week')['Sales'].sum()[0]
    tue=store_one.groupby('day_of_week')['Sales'].sum()[1]
    wed=store_one.groupby('day_of_week')['Sales'].sum()[2]
    thu=store_one.groupby('day_of_week')['Sales'].sum()[3]
    fri=store_one.groupby('day_of_week')['Sales'].sum()[4]
    
    
    day=(sat+sun+mon+tue+wed)/5
    end=(thu+fri)/2
    all_sales=(sat+sun+mon+tue+wed+thu+fri)
    dif=(end-day)/all_sales
    
    difs.append(dif)
```

```python
dict_dif=dict()
for i in range(1,55):
    dict_dif[i]=difs[i-1]
    
```
